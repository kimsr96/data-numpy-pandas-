```python
import requests # 크롤링에 사용하는 패키지
from bs4 import BeautifulSoup # html 변환에 사용함
import pandas as pd
import re

url = 'https://movie.naver.com/movie/running/premovie.nhn'
response = requests.get(url).text
html = BeautifulSoup(response, 'html.parser')

movies = html.find_all("dl", attrs={"class":"lst_dsc"})

title_list = []
link_list = []
up_list = []
down_list = []
genre_list = []
runningTime_list = []
releaseDate_list = []
for movie in movies:
    genres =[]
    if movie.find("dl", attrs={"class":"info_exp"}):
        title = movie.find("a").get_text().strip()
        link= movie.find("a")["href"]
        up = movie.find_all('em', attrs={"class":"exp_cnt"})[0].get_text()      
        down = movie.find_all('em', attrs={"class":"exp_cnt"})[1].get_text()
        genre = movie.find('span', attrs={"class":"link_txt"}).find_all('a', limit=3)
        
        for i in genre:
            genres.append(i.get_text())
        

        releaseDate = movie.find('dl', attrs={"class":"info_txt1"}).find('dd').get_text().replace("\t","").replace("\n","").replace("\r","").strip()
        print(releaseDate)
#         p = re.compile("분$")
#         m = p.match(releaseDate)
#         print(m.span())
        
        title_list.append(title)
        link_list.append(f"https://movie.naver.com"+link)
        up_list.append(up)
        down_list.append(down)
        genre_list.append(genres)
        releaseDate_list.append(releaseDate)



df = pd.DataFrame(columns=['title', 'link', 'up', 'down', 'genre','releaseDate'])
df['title']= title_list
df['link']= link_list
df['up']= up_list
df['down']=down_list
df['genre'] = genre_list
df['releaseDate']= releaseDate_list
# df['runningTime']= runningTime_list

# df.to_csv("movie.csv", mode='w')
```
